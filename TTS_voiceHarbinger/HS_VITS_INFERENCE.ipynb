```python
# Bibliotecas
import os
import locale
from google.colab import drive
```

```python
# Resolver questão de UTF-8 na geração do audio
locale.getpreferredencoding = lambda: "UTF-8"
```

```python
# Conectar ao Drive
drive.mount('/content/drive')
```

```python
mdl_path = 'checkpoint_1051500.pth'
cfg_path = '/'.join(mdl_path.split('/')[:-1]) + '/config.json'
chk_step = mdl_path.split('/')[-1].split('.')[0]
chk_step = 'best' if chk_step == 'best_model' else chk_step.split('_')[-1]
out_path3 =  'audio_skyrim_' + chk_step + '.wav'
fala3 = "Big Luan, Listen to me. I used to be an adventurer like you. Then I took an arrow in the knee."
```

```python
# Definir comando
comando3 = f'tts --text "{fala3}" --model_path {mdl_path} --config {cfg_path} --out_path {out_path3}'
```

```python
# Gerar audio
os.system(comando3)
```
